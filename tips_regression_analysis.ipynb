```python
# List to hold the accounts
accounts = []
# New account function
def create_account(account_number, initial_balance=0):
    account = {"account_number": account_number, "balance": initial_balance}
    accounts.append(account)
    print(f"Account {account_number} created with balance {initial_balance}.")
# Deposit money function
def deposit(account_number, amount):
    for account in accounts:
        if account['account_number'] == account_number:
            account['balance'] += amount
            print(f"Deposited {amount} into account {account_number}. New balance: {account['balance']}.")
            return account['balance']
    print("Account not found.")
    return

# Withdraw money function
def withdraw(account_number, amount):
    for account in accounts:
        if account['account_number'] == account_number:
            if account['balance'] >= amount:
                account['balance'] -= amount
                print(f"Withdrew {amount} from account {account_number}. New balance: {account['balance']}.")
                return account['balance']
            else:
                print("Insufficient balance.")
                return
    print("Account not found.")
    return

# Check balance function
def check_balance(account_number):
    for account in accounts:
        if account['account_number'] == account_number:
            print(f"Account {account_number} balance: {account['balance']}.")
            return account['balance']
    print("Account not found.")
    return

# Main menu
def Menu():
    print("Welcome to the Banking System!")
    while True:
        print("\nSelect an option:")
        print("1. Create a new account")
        print("2. Deposit money")
        print("3. Withdraw money")
        print("4. Check account balance")
        print("5. Exit")

        choice = input("Enter choice (1-5): ")

        if choice == '1':
            account_number = input("Enter account number: ")
            initial_balance = float(input("Enter initial balance: "))
            create_account(account_number, initial_balance)

        elif choice == '2':
            account_number = input("Enter account number for deposit: ")
            amount = float(input("Enter deposit amount: "))
            deposit(account_number, amount)

        elif choice == '3':
            account_number = input("Enter account number for withdrawal: ")
            amount = float(input("Enter withdrawal amount: "))
            withdraw(account_number, amount)

        elif choice == '4':
            account_number = input("Enter account number to check balance: ")
            check_balance(account_number)

        elif choice == '5':
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

Menu()
```
