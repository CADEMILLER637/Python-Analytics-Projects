```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/dc_bike_share-1.csv')
print(df.head())
print()
# Create and use a histogram to explore and better understand the distribution of the number of registered bike users
plt.hist(df['registered'], bins=30)
plt.title('Distribution of Registered Bike Riders')
plt.ylabel('Frequency')
plt.xlabel('Number of Registered Riders')
plt.show()

# Business insights
print('The large frequency of hours with 0-30 riders indicates that there is significant downtime')
# Median number of registered riders (grouped by month) for each month for the year 2011.
df_2011 = df[df['yr'] == 0]
df_2012 = df[df['yr'] == 1]

median_2011 = df_2011.groupby('mnth')['registered'].median()
median_2012 = df_2012.groupby('mnth')['registered'].median()
plt.bar(median_2011.index, median_2011.values)
plt.title('Median Number of Registered Riders by Month 2011')
plt.ylabel('Median Number of Registered Riders')
plt.xlabel('Month')
plt.show()
#Business insights
print('The peak of monthly users is in june, while the lowest point is january.')

# Create a second bar plot of the same data for 2012.
plt.bar(median_2012.index, median_2012.values)
plt.title('Median Number of Registered Riders by Month (2012)')
plt.ylabel('Median Number of Registered Riders')
plt.xlabel('Month')
plt.show()
#Business insights
print('The peak of monthly users is in September, while the lowest point is january.')

#Create a bar plot showing the median number of registered riders for each hour for the month of July
df_july = df[df['mnth'] == 7]
median_july = df_july.groupby('hr')['registered'].median()
plt.bar(median_july.index, median_july.values)
plt.title('Median Number of Registered Riders by Hour (July, 2011-2012)')
plt.ylabel('Median Number of Registered Riders')
plt.xlabel('Hour of the Day')
plt.show()
#Business insights
print('There are major spikes in sales at 8am, and from 5-6pm.')

# Create a bar plot showing the median number of registered riders (grouped by day) for each day (include both years)
median_weekday = df.groupby('weekday')['registered'].median()

plt.bar(median_weekday.index, median_weekday.values)
plt.title('Median Number of Registered Riders by Day of the Week (2011-2012)')
plt.ylabel('Median Number of Registered Riders')
plt.xlabel('Day of the Week')
plt.show()
#Business insights
print('the busiest day is friday, while the least busy day is sunday')

#Create a scatter plot to show the relationship between windspeed and the number of registered riders only for the month of March in the year 2011.
df_march_2011 = df[(df['yr'] == 0) & (df['mnth'] == 3)]

plt.scatter(df_march_2011['windspeed'], df_march_2011['registered'], alpha=0.7, color='brown')
plt.title('Relationship Between Windspeed and Registered Riders (March 2011)')
plt.xlabel('Windspeed')
plt.ylabel('Number of Registered Riders')
plt.show()
# Business insights
print('The prefered wind speed seems to be from 0-0.4')

# One business decision or objective
# Increase bike availability during peak hours and reduce availability during slow hours
```
