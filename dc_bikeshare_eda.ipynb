```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/purchase_data.csv')

print(df.head())
print()
print(df.describe())
print()
# Creating a plot to see the gender distrobution fo the data filtered by income
df_filtered = df[df['Income'] > 55852.24]
sns.countplot(data=df_filtered, x='Gender')
plt.title('Gender Distrobution by Income Under the Median')
plt.show()

# Creating a histogram to see the distrobution for the Purchase amounts
plt.hist(df['Purchase_Amount'], bins=20)
plt.title('Purchase Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
print()

# creating a heatmap of the correlation matrix of the data
df_numeric = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = df_numeric.corr()
print(correlation_matrix)
print()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()

avg_purchase_by_age = df.groupby('Age')['Purchase_Amount'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_purchase_by_age, x='Age', y='Purchase_Amount')
plt.title('Average Purchase Amount by Age')
plt.xlabel('Age')
plt.ylabel('Average Purchase Amount')
plt.xticks(rotation=45)
plt.show()

avg_rating_by_age = df.groupby('Age')['Product_Rating'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_rating_by_age, x='Age', y='Product_Rating')
plt.title('Average Product Rating by Age')
plt.xlabel('Age')
plt.ylabel('Average Product Rating')
plt.xticks(rotation=45)
plt.show()
```
